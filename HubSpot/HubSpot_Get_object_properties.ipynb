## Input

### Import library

```python
# from naas_drivers import hubspot
import naas
import requests
import pandas as pd
from pprint import pprint
```

### Setup your HubSpot
ðŸ‘‰ Access your [HubSpot API key](https://knowledge.hubspot.com/integrations/how-do-i-get-my-hubspot-api-key)

```python
HS_API_KEY = naas.secret.get("HS_API_KEY") or 'YOUR_HUBSPOT_API_KEY'
```

## Model

### Get single deal

```python
def get_properties(obj):
    url = f"https://api.hubapi.com/crm/v3/properties/{obj}/"
    querystring = {
        "archived": "false",
        "hapikey": HS_API_KEY,
    }
    headers = {'accept': 'application/json'}
    # Requests data
    res = requests.get(url,
                       headers=headers,
                       params=querystring)
    res.raise_for_status()
    return res.json()

properties = get_properties("contacts")
```

## Output

### Display result

```python
df_properties = pd.DataFrame()

for p in properties.get("results"):
    # Manage meta
    meta = p.get("modificationMetadata")
    p.pop("modificationMetadata")
    p.update(meta)
    # Manage options
    options = p.get("options")
    p.pop("options")
    
    data = []
    for option in options:
        p.update(option)
        data.append(p)
    break

pd.DataFrame(data)
```
